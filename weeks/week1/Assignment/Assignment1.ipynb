**8. Take a positive integer n as input and print all prime numbers from 2 to n.**

```python
n=int(input("enter the number to check whether it is prime or not"))
for num in range(2,n+1):
    is_prime=True
    for i in range(2,num):
        if num % i ==0:
            is_prime=False
        break
    if is_prime:
        print(num)
```

**9. Keep accepting numbers from the user until they enter 0. Then print the total count of
numbers entered, their sum, and their average. Do not include 0 in the calculations.**

```python
count = 0
total = 0

num = int(input("Enter a number (0 to stop): "))

while num != 0:
    total = total + num
    count = count + 1
    num = int(input("Enter a number (0 to stop): "))

if count != 0:
    print("Count =", count)
    print("Sum =", total)
    print("Average =", total / count)
else:
    print("No numbers entered")
    
```

**10. Take n as input and print a triangle pattern of n rows made of asterisks, where row i
has (2i - 1) stars centered appropriately.**

```python
n = int(input("Enter number of rows: "))

for i in range(1, n + 1):
    print(" " * (n - i) + "*" * (2 * i - 1))
```
