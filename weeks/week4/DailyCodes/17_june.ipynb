```python
# else block ---> if try block doesnt raise exception else block will be executed
```

```python
try:
    f=open('a1.txt')
except FileNotFoundError:
    print("file does not exist")
except Exception as e:
    print(e)

else:
    print(f.read())
    f.close()
```

```python
# finally --->  this will be ecectued dispite any error occurs
```

```python
try:
    f=open('a2.txt')
except FileNotFoundError:
    print("file does not exist")
except Exception as e:
    print(e)

else:
    print(f.read())
    f.close()
finally:
    print("this is finally block no one can stop me from executing -----> i am a don")
```
