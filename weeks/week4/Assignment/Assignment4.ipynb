1. write a function safe_divide(a,b) that divides two umbers and handles ZeroDivisionError and TypeError separately, printing an appropriate message for each. use a finally block to print 'opertaional attempted' regardless of the outcome. test it with valid numbers, division by zero, and a string input.

```python
def safe_division(a,b):
    try:
        result=a/b
        print("result is",result)
    except ZeroDivisionError:
        print("Error: cannot divide by zero ")
    except TypeError:
        print("Invaid input type")
    finally:
        print("Operation attempted")
```

```python
# valid numbers
safe_division(6,3)
```

```python
#division by zero
safe_division(10,0)
```

```python
#type error
safe_division(10,'2')
```

2. write a function validate_age(age) that raises a custom exception InvalidAgeError (a class you define, inheriting from Exception) if age is negative or greater than 120. Catch this exception in a calling block and print a user-friendly message. Test it with at least 3 different values.

```python
class InvalidAgeError(Exception):
    pass
def validate_age(age):
    if age<0:
        raise InvalidAgeError("Age cannot be negative")
    elif age>120:
        raise InvalidAgeError("Age cannot be greater than 120")
    else:
        print("valid age",age)
test_age=[-25,18,150,25]
for age in test_age:
    try:
        validate_age(age)
    except InvalidAgeError as e:
        print(f"Invalid age entered({age}),{e}")
        
```

6.Generate a NumPy array of 15 random integers between 1 to 100 using np.randoom.randint() (set a seed of 42 for reproducibility ) than:
a. find the mean,median,standard deviation,min,max 
b. sort the array in descending order 
c. find the indices of the 3 largest values using np.argsort()
d.Normalize the array so all values lie between 0 and 1 

```python
import numpy as np
np.random.seed(42)
arr = np.random.randint(1, 101, 15)

print("Original Array:",arr)
print("\nStatistics:")
print("Mean =", np.mean(arr))
print("Median =", np.median(arr))
print("Standard Deviation =", np.std(arr))
print("Minimum =", np.min(arr))
print("Maximum =", np.max(arr))
desc_arr = np.sort(arr)[::-1]

print("\nArray sorted in descending order:")
print(desc_arr)

largest_indices = np.argsort(arr)[-3:][::-1]

print("\nIndices of the 3 largest values:")
print(largest_indices)

print("The 3 largest values are:")
print(arr[largest_indices])
normalized_arr = (arr - np.min(arr)) / (np.max(arr) - np.min(arr))
print("\nNormalized Array:")
print(normalized_arr)
```

7.create a Pandas DataFrame from the dictionary below and perform the following:
data={"Name":["Aarav","priya","Karan","Meera","Sahil"],"Age":[23,21,25,22,24],"Department":["AI","Data Science","AI","Web Dev","Data Science"],"Score":[88,92,79,95,67],}
a.Display the frist 3 rows and the data types of each column.
b. Add a new column Grade based on Score(>=90'A',>=75'B',else'c')
c.sort the DataFrame by score in descending order
d.filter rows where Deparatment is 'AI'


```python
import pandas as pd
data = {
    "Name": ["Aarav", "priya", "Karan", "Meera", "Sahil"],
    "Age": [23, 21, 25, 22, 24],
    "Department": ["AI", "Data Science", "AI", "Web Dev", "Data Science"],
    "Score": [88, 92, 79, 95, 67]
}
df = pd.DataFrame(data)
print("First 3 rows:")
print(df.head(3))

print("\nData types:")
print(df.dtypes)
def get_grade(score):
    if score >= 90:
        return 'A'
    elif score >= 75:
        return 'B'
    else:
        return 'C'

df["Grade"] = df["Score"].apply(get_grade)

print("\nDataFrame with Grade:")
print(df)
df_sorted = df.sort_values(by="Score", ascending=False)

print("\nSorted by Score (descending):")
print(df_sorted)
df_ai = df[df["Department"] == "AI"]

print("\nDepartment = AI:")
print(df_ai)
```

```python
import pandas as pd

data = {
    "Name": ["Aarav", "priya", "Karan", "Meera", "Sahil"],
    "Age": [23, 21, 25, 22, 24],
    "Department": ["AI", "Data Science", "AI", "Web Dev", "Data Science"],
    "Score": [88, 92, 79, 95, 67]
}

df = pd.DataFrame(data)
dept_avg = df.groupby("Department")["Score"].mean()
print(dept_avg)
max_index = df["Score"].idxmax()
highest_row = df.loc[max_index]
print(highest_row)
df_renamed = df.rename(columns={"Score": "Marks"})
print(df_renamed)
df_dropped = df.drop(columns=["Age"])
print(df_dropped)
```

9. Create a DataFrame with some missing values as shown below and handle thm:
import numpy as np data={"product":["pen","Notebook","Eraser","sharpner","Ruler"],"Price":[10,np.nan,5,8,np.nan],"Stock":[100,50,np.nan,30,20]}
a. check the number of missing values in each column using isnull().sum()
b.Fill missing price values with the column mean
c Drop rows where stock is still missing
d.Save the cleaned DataFRame to a CSV file named Cleaned_products.csv without the index

```python
import numpy as np
import pandas as pd

data = {
    "product": ["pen", "Notebook", "Eraser", "sharpner", "Ruler"],
    "Price": [10, np.nan, 5, 8, np.nan],
    "Stock": [100, 50, np.nan, 30, 20]
}

df = pd.DataFrame(data)
print("Missing values in each column:")
print(df.isnull().sum())
df["Price"] = df["Price"].fillna(df["Price"].mean())
df = df.dropna(subset=["Stock"])
df.to_csv("Cleaned_products.csv", index=False)
print("\nCleaned DataFrame:")
print(df)
```
