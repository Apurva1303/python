```python
df=pd.read_csv('satandar.csv',nrows=40000)
```

```python
df.shape
```

```python
print(df.isnull().sum())
```

```python
df.columns
```

```python
from sklearn.feature_selection import VarianceThreshold
```

```python
vt=VarianceThreshold(threshold=0.5)
```

```python
vt.fit(df)
```

```python
vt.get_support()
#true==>these are the columns which we will be keeping
#false--> remoove these columns
#filter out the columns
```

```python
new=df.loc[:,vt.get_support()]
```

```python
new.shape
```

```python
#from this new dataset ... chcek if there are duplicated columns and remove them
```

```python
new.duplicated().sum()
#by default it will act on rows
```

```python
df=pd.read_csv('mtcars.csv')
df
```

```python
df.columns
```

```python
x=df[['disp','cyl','wt']]
```

```python
y=df['mpg']
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
lr.fit(x,y)
```

```python
lr.score(x,y) #r2 score
```

```python
#RFE
```

```python
from sklearn.feature_selection import RFE
```

```python
rfe=RFE(LinearRegression(), n_features_to_select=3)
```

```python
x=df.drop('mpg',axis=1)
```

```python
rfe.fit(x,y)
```

```python
rfe.get_feature_names_out()
```

```python
rfe.get_support()
```

```python
x_new=x.loc[:,rfe.get_support()]
x_new.columns
```
