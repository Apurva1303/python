```python
2: Sum of Even Numbers
Write Python code that iterates through the following list:numbers = [12, 7, 23, 48, 15, 66, 9]
Calculate and print the sum of all even numbers in the list.
```

```python
numbers = [12, 7, 23, 48, 15, 66, 9]

total = 0

for num in numbers:
    if num % 2 == 0:
        total = total + num

print("Sum of even numbers =", total)
```

```python
3.Manual List Reversal
Given a list data = ['apple', 'banana', 'cherry', 'date', 'elderberry']
create a new list that contains the elements of data in reverse order.

You are not allowed to use data.reverse() or data[::-1].

You must achieve this using loops and basic list operations.
```

```python
data = ['apple', 'banana', 'cherry', 'date', 'elderberry']

reverse_list = []

for item in data:
    reverse_list.insert(0, item)

print(reverse_list)
```

```python
4.Finding the Maximum Element
Without using the built-in max() function, write code to find and print the largest number in the list:values = [8, 2, 10, 5, 17, 3, 14]
```

```python
values = [8, 2, 10, 5, 17, 3, 14]

largest = values[0]

for num in values:
    if num > largest:
        largest = num

print("Largest number =", largest)
```

```python
5.Count Occurrences
Given the list items = ['red', 'blue', 'green', 'red', 'yellow', 'blue', 'red']
write code to count how many times the string 'red' appears in the list and print the count. Do not use the list.count() method.
```

```python
items = ['red', 'blue', 'green', 'red', 'yellow', 'blue', 'red']

count = 0

for color in items:
    if color == 'red':
        count = count + 1

print("Count of red =", count)
```

```python
6.Remove All Occurrences of an Element
Given a list mixed_numbers = [1, 5, 2, 5, 3, 5, 4, 5],write code to remove all instances of the number 5 from the list. The final list should only contain [1, 2, 3, 4].
You must achieve this by iterating and creating a new list or modifying the existing one carefully.
```

```python
mixed_numbers = [1, 5, 2, 5, 3, 5, 4, 5]

new_list = []

for num in mixed_numbers:
    if num != 5:
        new_list.append(num)

print(new_list)
```

```python
Question 7: List Palindrome Check
```

```python
list1 = [1, 2, 3, 2, 1]
list2 = ['a', 'b', 'c', 'b', 'd']
for lst in [list1, list2]:
    rev = []
    for i in range(len(lst)-1, -1, -1):
        rev.append(lst[i])
    if lst == rev:
        print("Palindrome")
    else:
        print("Not a Palindrome")
```

```python
Question 8: Merge and Sort Two Lists (Manual Sort)
```

```python
list_a = [3, 1, 4]
list_b = [2, 5, 0]
combined = list_a + list_b
for i in range(len(combined)):
    for j in range(len(combined) - 1):
        if combined[j] > combined[j + 1]:
            temp = combined[j]
            combined[j] = combined[j + 1]
            combined[j + 1] = temp
print("Sorted List:", combined)
```

```python
Question 9: Flaen a Nested List
```

```python
nested_list = [1, [2, 3], 4, [5, 6, 7], 8]

flat_list = []

for i in nested_list:
    if type(i) == list:
        for j in i:
            flat_list.append(j)
    else:
        flat_list.append(i)

print(flat_list)
```

```python
list_p = [1, 2, 3, 4, 5]
list_q = [4, 5, 6, 7, 8]
common = []
for i in list_p:
    for j in list_q:
        if i==j:
            common.append(i)
print("Common elements:",common) 
```
