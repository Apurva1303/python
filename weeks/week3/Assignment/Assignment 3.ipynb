```python
class BankAccount:
    def __init__(self,acc_holder,acc_number,acc_balance=0):
        self.acc_holder=acc_holder
        self.acc_number=acc_number
        self.acc_balance=acc_balance
    def deposit(self,amount):
        if amount>0:  #check if the amount is greater then zero
            self.acc_balance +=amount # add the new aount on right and stores the result back in the variable on left
            print(f"${amount} withdrawn sucessfully")
        else:
            print("not successfull")
    def withdraw(self,amount):
        if amount<=self.acc_balance:  #check whether your account has enough money
            self.acc_balance -=amount  # if yes then it will reduce the amount of money
            print(f"${amount} withdrawn successfully")
        else:
            print("apna balance jitna hai utna amount daloooo")
    def get_balance(self):
        return self.balance

    def __str__(self):  #will provide a readable description of the object in the form of string
        return (f"Account Holder:{self.acc_holder}\n"
                f"Account number:{self.acc_number}\n"
                f"Account Balance:${self.acc_balance:.2f}")
        #:.2f -> format specifier  
        # f -> display the number as a floating-point (decimal) number.
        #.2 -> show exactly 2 digits after the decimal point
        
```

```python
# Getters -> is a method used to get (read) the values of a private attributes.
# Setter -> is a method used to set (change) te value of a private attributes.
```

```python
class Employee:
    def __init__(self, name, emp_id, salary):
        self.__name=name
        self.__emp_id=emp_id
        self.__salary=salary

        # Getters
    def get_name(self):
        return self.__name

    def get_emp_id(self):
        return self.__emp_id

    def get_salary(self):
        return self.__salary

    # setters
    def set_name(self,name):
        self.__name=name
    def set_emp_id(self,emp_id):
        self.__emp_id=emp_id
    def set_salary(self,salary):
        if salary <10000:
            # ValueError is a built-in Python error that occurs when:
#The value given to a function is of the correct type, but the value itself is not acceptable.
            raise ValueError("salary cannot be below 10000")
        self.__salary=salary

        # Raise salary
    def give_raise(self, percent):
        self.__salary+=self.__salary*percent /100
```

```python
5. Create a base class Shape with a method area() that returns 0 and a method describe() that prints 'i am a shape'. create three subclasses:
 a. Circle(radius) -- overrides area() using pi* r^2.
 b. Rectangle(length,width) -- overrides area()
 c. Triangle(base,height) -- overrides area()
Each subclass should also override describe() to print something specific about itself. Create one object of each and call both methods 
```

```python
import math
class Shape:
    def area(self):
        return 0
    def describe(self):
        print("i am a shape")
class Circle (Shape):
    def __init__(self,radius):
        self.radius=radius
    def area(self):
        return math.pi*self.radius*self.radius
    def describe(self):
        print("i am a Circle")


class Rectangle (Shape):
    def __init__(self,length,width):
        self.length=length
        self.width=width
    def area(self):
        return self.length*self.width
    def describe(self):
        print("i am a Ractangle")

class Triangle(Shape):
    def __init__(self,base,height):
        self.base=base
        self.height=height
    def area(self):
        return 0.5* self.base* self.height
    def describe(self):
        print("i am a Triangle")
        
```

```python
c=Circle(5)
r=Rectangle(2,3)
t=Triangle(3,8)
```

```python
c.describe()
print("area:-", c.area())
print()
```

```python
r.describe()
print("area:-", r.area())
print()
```

```python
t.describe()
print("area:-", t.area())
```

```python
6. Create a class Vehicle with attributes make, model, and year and a method start_engine() that prints 'engine started'. Create two subclasses:
 a.Car(make,model,year,num_doors)--overrides start_engine() to include the cars's name in the message
 b. ElectricCar(make,model,year,battery_capacity)--overrides start_engine() to print 'silent start -battery at X kWh'.
 Use super() in both subclasses to call parent __init__. Create objects and test all methods   
```

```python
# Parent class
class Vehicle:
    def __init__(self, make, model, year):
        self.make = make
        self.model = model
        self.year = year

    def start_engine(self):
        print("Engine started")


# Subclass Car
class Car(Vehicle):
    def __init__(self, make, model, year, num_doors):
        super().__init__(make, model, year)  # Call parent constructor
        self.num_doors = num_doors

    def start_engine(self):
        print(f"Engine started for {self.make} {self.model}")


# Subclass ElectricCar
class ElectricCar(Vehicle):
    def __init__(self, make, model, year, battery_capacity):
        super().__init__(make, model, year)  # Call parent constructor
        self.battery_capacity = battery_capacity

    def start_engine(self):
        print(f"Silent start - battery at {self.battery_capacity} kWh")
```

```python
# Creating objects
vehicle = Vehicle("Honda", "Civic", 2020)
car = Car("Toyota", "Corolla", 2022, 4)
electric_car = ElectricCar("Tesla", "Model 3", 2024, 75)
```

```python

class Animal:
    def sound(self):
        print("Some sound")

class Dog(Animal):
    def sound(self):
        print("Bark")

class Cat(Animal):
    def sound(self):
        print("Meow Meoww!")

class Cow(Animal):
    def sound(self):
        print("Moo MOO!")        
```

```python
# Function demonstrating polymorphis
def make_all_sounds(animals):
    for animal in animals:
        animal.sound()
```

```python
# Creating objects
dog = Dog()
cat = Cat()
cow = Cow()
```

```python
# List of animal objects
animals = [dog, cat, cow]
```

```python
# Calling function
make_all_sounds(animals)
#Due to polymorphism, the correct overridden method is executed based on the actual object type (Dog, Cat, or Cow), even though all objects are treated as Animal references.
```

```python
# Calculator class
class Calculator:
    def add(self, a, b, c=0):
        return a + b + c

    def multiply(self, a, b, c=1):
        return a * b * c


# Vector class
class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    # Overloading + operator
    def __add__(self, other):
        return Vector(self.x + other.x, self.y + other.y)

    def __str__(self):
        return f"({self.x}, {self.y})"
```

```python

# Testing Calculator
calc = Calculator()

print("Addition of 2 numbers:", calc.add(10, 20))
print("Addition of 3 numbers:", calc.add(10, 20, 30))

print("Multiplication of 2 numbers:", calc.multiply(2, 3))
print("Multiplication of 3 numbers:", calc.multiply(2, 3, 4))
```

```python
10. Built a menu_driven Student Record Manager using OOP and file handlig. The app should have a class studentsManager with the following menu options:
  a. Add Student - takes name,roll number,and marks for 3 subjects.
  b. view All-sidplays all stored students with their average marks.
   c. Serch student - finds a student by roll number
   d. Delete students -removes a student by roll number
   e. save to file- writes all records to students_data.txt
   f. load from file- reads records back from students_data.txt on startup.
    g. Exit

       Requirments: use a class with a list as internal storgae, handle invalid menu choices gracefully, and use a while loop to keep the menu running until the user chooses to exit
```

```python
class StudentsManager:
    def __init__(self):
        self.students = []
        self.load_from_file()

    # a. Add student
    def add_student(self):
        name = input("Enter name: ")
        roll = input("Enter roll no: ")

        marks1 = float(input("Enter marks 1: "))
        marks2 = float(input("Enter marks 2: "))
        marks3 = float(input("Enter marks 3: "))

        self.students.append([name, roll, marks1, marks2, marks3])
        print("Student added!")

    # b. View all
    def view_all(self):
        if len(self.students) == 0:
            print("No records found")
            return

        for s in self.students:
            avg = (s[2] + s[3] + s[4]) / 3
            print(f"\nName: {s[0]}")
            print(f"Roll: {s[1]}")
            print(f"Average: {avg:.2f}")

    # c. Search
    def search_student(self):
        roll = input("Enter roll no: ")

        for s in self.students:
            if s[1] == roll:
                print("Student Found!")
                print("Name:", s[0])
                print("Marks:", s[2], s[3], s[4])
                return

        print("Not found")

    # d. Delete
    def delete_student(self):
        roll = input("Enter roll no: ")

        for s in self.students:
            if s[1] == roll:
                self.students.remove(s)
                print("Deleted successfully")
                return

        print("Not found")

    # e. Save to file
    def save_to_file(self):
        file = open("students_data.txt", "w")

        for s in self.students:
            file.write(s[0] + "," + s[1] + "," +
                       str(s[2]) + "," + str(s[3]) + "," + str(s[4]) + "\n")

        file.close()
        print("Saved to file")

    # f. Load from file
    def load_from_file(self):
        try:
            file = open("students_data.txt", "r")
            self.students = []

            for line in file:
                data = line.strip().split(",")

                self.students.append([
                    data[0],
                    data[1],
                    float(data[2]),
                    float(data[3]),
                    float(data[4])
                ])

            file.close()
        except:
            self.students = []


# g. MENU
manager = StudentsManager()

while True:
    print("\n1.Add 2.View 3.Search 4.Delete 5.Save 6.Exit")
    choice = input("Enter choice: ")

    if choice == "1":
        manager.add_student()

    elif choice == "2":
        manager.view_all()

    elif choice == "3":
        manager.search_student()

    elif choice == "4":
        manager.delete_student()

    elif choice == "5":
        manager.save_to_file()

    elif choice == "6":
        print("Exiting...")
        break

    else:
        print("Invalid choice")
```
