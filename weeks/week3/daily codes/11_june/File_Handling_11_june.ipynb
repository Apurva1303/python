```python
#write into a myfile.txt with context manager
#use write function and add 3 separate lines
```

```python
with open('myfile.txt', 'w') as file:
    file.write("Maharashtra is a beautiful state.\n")
    file.write("Mumbai is its capital city.\n")
    file.write("Marathi is the official language.\n")

print("Data written successfully.")
```
