```python
class Employee:
    company='wipro'

    def __init__(self,name,mon_sal,job_dest):
        self.name=name
        self.mon_sal=mon_sal
        self.job_dest=job_dest
    def annual_sal(self):
        print("annual sal",self.mon_sal*12)
    def info(self):
        return f"name=>{self.name},company=>{self.company},mon_sal=>{self.mon_sal},job_dest{self.job_dest}"
    
        
        
```

```python
s1=Employee('Apurva',12000,'hr')
s2=Employee('Sukanya',13400,'manager')
s3=Employee('Meera',12000,'web designer')
```
