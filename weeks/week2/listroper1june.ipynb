```python
Characteristics of list
1.list are hetrogeneous and also mutable and dublicates are alllowes
2.list are ordered
3.list slicing
  list[start:stop:step]
    start → index where slicing begins (inclusive)
    stop → index where slicing ends (exclusive)
    step → how much to jump between elements
    *here the last stop index element is skipped
4.list can be reversed
5. list supports membership operator
6.Arithematic operators can aslo be supported
```
