1. Write a function analyze_sentence(sentence) that takes a sentence and returns a dictionary with
word_count,
char_count (excluding spaces),
unique_words (case-insensitive),
and most_frequent_word.
    Test it with:"The quick brown fox jumps over the lazy dog the fox"

```python
def analyze_sentence(sentence):
    words = sentence.lower().split()

    return {
        "word_count": len(words),
        "char_count": len(sentence.replace(" ", "")),
        "unique_words": len(set(words)),
        "most_frequent_word": max(words, key=words.count)
    }
```

```python
analyze_sentence("The quick brown fox jumps over the lazy dog the fox")
```

2. Given the string below, do the following and print each result:
text = "A man a plan a canal Panama"
a) Reverse the string without using slicing.
b) Check if it is a palindrome, ignoring case and spaces.
c) Replace every vowel with *.
d) Count how many times each vowel appears.

```python
text = "A man a plan a canal Panama"

# a) Reverse the string
rev = ""
for i in text:
    rev = i + rev
print("Reverse:", rev)

# b) Palindrome check
s = text.replace(" ", "").lower()
r = ""
for i in s:
    r = i + r
print("Palindrome:", s == r)

# c) Replace vowels with *
new = ""
for i in text:
    if i.lower() in "aeiou":
        new += "*"
    else:
        new += i
print("After replacing vowels:", new)

# d) Count vowels
print("a:", text.lower().count("a"))
print("e:", text.lower().count("e"))
print("i:", text.lower().count("i"))
print("o:", text.lower().count("o"))
print("u:", text.lower().count("u"))
```

```python
def format_report(data):
    print(f"{'Name':<15}{'Score':>6}{'Grade':^8}")
  

    for name, score, grade in data:
        print(f"{name:<15}{score:>6}{grade:^8}")

data = [
    ("Alice", 92, "A"),
    ("Bob", 78, "B+"),
    ("Charlie", 65, "C"),
    ("Diana", 100, "A+")
]

format_report(data)
```

```python
4. Write a function list_ops(nums) that:
a) Removes duplicates while preserving the original order.
b) Returns a new list with only elements at even indices, squared.
c) Flattens the nested list into a single list using list comprehension.nums = [3, 1, 4, 1, 5, 9, 2, 6, 5, 3] nested = [[1, 2, 3], [4, 5], [6, 7, 8, 9]]
```

```python
def list_ops(nums, nested):

 
    unique = []
    for num in nums:
        if num not in unique:
            unique.append(num)


    squared_even = [nums[i]**2 for i in range(0, len(nums), 2)]


    flat = [item for sublist in nested for item in sublist]

    return unique, squared_even, flat


nums = [3, 1, 4, 1, 5, 9, 2, 6, 5, 3]
nested = [[1, 2, 3], [4, 5], [6, 7, 8, 9]]

unique, squared_even, flat = list_ops(nums, nested)

print("Without duplicates:", unique)
print("Even-index elements squared:", squared_even)
print("Flattened list:", flat)
```

7.Write a function inventory_manager(records) that takes a list of tuples
(product, category, quantity, price) and returns a dictionary grouped by category.

Each category should have:
- total_items
- total_value
- a list of product names

Use:

records = [
    ("Laptop", "Electronics", 10, 55000),
    ("Mouse", "Electronics", 50, 499),
    ("Desk", "Furniture", 8, 12000),
    ("Chair", "Furniture", 15, 4500),
    ("Pen", "Stationery", 200, 20)
]

```python
def inventory_manager(records):
    inventory = {}

    for product, category, quantity, price in records:

        if category not in inventory:
            inventory[category] = {
                "total_items": 0,
                "total_value": 0,
                "products": []
            }

        inventory[category]["total_items"] += quantity
        inventory[category]["total_value"] += quantity * price
        inventory[category]["products"].append(product)

    return inventory


```

```python
records = [
    ("Laptop", "Electronics", 10, 55000),
    ("Mouse", "Electronics", 50, 499),
    ("Desk", "Furniture", 8, 12000),
    ("Chair", "Furniture", 15, 4500),
    ("Pen", "Stationery", 200, 20)
]

result = inventory_manager(records)

for category, details in result.items():
    print(category, ":", details)
```

```python
def power_all(*args, exp=2):
    result = []

    for i in args:
        result.append(i ** exp)

    return result

print(power_all(2, 3, 4))
```

```python
10. Complete the following:

a) Write a closure make_multiplier(n) that returns a function
multiplying its input by n.
Create triple and quintuple using it.

b) Write a decorator @timer that prints the execution time
of any function it wraps.
Apply it to slow_sum(n) which computes the sum of squares
from 1 to n.

c) Write a decorator @validate_positive that raises a
ValueError if any argument is zero or negative.
Apply it to safe_sqrt(x).
```

```python
def make_multiplier(n):

    def multiply(x):
        return x * n

    return multiply


triple = make_multiplier(3)
quintuple = make_multiplier(5)

print(triple(10))
print(quintuple(10))
```

```python
import time

def timer(func):

    def wrapper(n):
        start = time.time()

        result = func(n)

        end = time.time()

        print("Time Taken =", end - start)

        return result

    return wrapper


@timer
def slow_sum(n):
    total = 0

    for i in range(1, n + 1):
        total += i * i

    return total


print(slow_sum(1000))
```

```python
import math

def validate_positive(func):

    def wrapper(x):
        if x <= 0:
            raise ValueError("Number should be positive")

        return func(x)

    return wrapper


@validate_positive
def safe_sqrt(x):
    return math.sqrt(x)


print(safe_sqrt(25))
```
