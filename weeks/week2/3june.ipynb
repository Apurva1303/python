```python
"""take a string as user input
use a loop and if else to checkk if a specific character is present in the string or not.
get output for both if the character is present and if it is not present.
    take 2 character to check in the string
    """
```

```python
ss=input("enter a string")
char1=input("enter the frist char to check")
char2=input("enter the second char to check")

fount_char1=False
found_char2=False

for char in ss:
    if char==char1:
        found_char1=True
    if char==char2:
        found_char2=True
if found_char1:
    print(f"the character'{char1}' is present in the string")
else:
    print(f"the character'{char1}' is not present in the string")
if found_char2:
    print(f"the character'{char2}' is present in the string")
else:
    print(f"the character'{char2}' is not present in the string")
        
    
```
