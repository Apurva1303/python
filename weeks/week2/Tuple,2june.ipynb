#set operation

```python
s6={2,3,4,5,6,7}
s5={12,3,4,23,54,66}
```

```python
s6.union(s5) # return all elements duplicate will be removed
```

```python
s6 | s5  # another method of union
```

```python
s6.intersection(s5) # it will return the common values from both sets
```

```python
s6.difference(s5) 
# it will return the vales which are differnt from set 1 only
```

```python
s6.symmetric_difference(s5)
# it will return the vales which are differnt from both sets
```

```python
sA={1,2,3,4,5,6,7}
sB={4,5,6,7}
sC={1,2,3}
```

```python
sB.issubset(sA)
```

```python
sA.issuperset(sB)
```

```python
sC.issubset(sB)
```
