```python
s = input("Enter a sentence: ")

words = s.split()

short = []
medium = []
long = []

for w in words:
    if len(w) < 5:
        short.append(w)
    elif len(w) <= 8:
        medium.append(w)
    else:
        long.append(w)

print("Short:", short)
print("Medium:", medium)
print("Long:", long)
```

```python
lst = ["python", "pythonthon", "pycham","pynumthon", "java"]

result = []

for i in lst:
    if i.startswith("py") and i.endswith("thon"):
        result.append(i)

print(result)
```

```python
s = input("Enter paragraph: ")

s = s.lower()

for ch in ".,!?":
    s = s.replace(ch, "")

words = s.split()

unique = set()

for w in words:
    unique.add(w)

print(sorted(unique))
```

```python
s = input("Enter string: ")

even = []
odd = []

for i, ch in enumerate(s):
    if i % 2 == 0:
        even.append(ch)
    else:
        odd.append(ch)

print("Even:", even)
print("Odd:", odd)
```

```python
s1 = input("Enter first string: ")
s2 = input("Enter second string: ")

l1 = sorted(list(s1))
l2 = sorted(list(s2))

if l1 == l2:
    print("Anagram")
else:
    print("Not Anagram")
```

```python
letters = set("abcdefghijklmnopqrstuvwxyz")

s = input("Enter string: ").lower()

for ch in s:
    if ch in letters:
        letters.remove(ch)

print("Missing letters:")
print(letters)
```
