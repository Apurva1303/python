```python
#function to check which number is larger
def large(a,b):
    if (a>b):
        print("A is larger",a)
    elif (b>a):
        print("B is larger",b)
    else:
        print("both are equal")
```

```python
large(4,4)
```

```python
#filter and map
#filter-> it will only filter the vaules from the data structure
#map-> it will change the values of a data structure
l=[1,2,3,4,5,6,7,8]
```

```python
n1=filter(lambda x:x%2==0, l)
list(n1)
```

```python
n2=map(lambda x:x*x,l)
list(n2)
```

```python
n1=map(lambda x:x%2==0, l)
list(n1)
```
