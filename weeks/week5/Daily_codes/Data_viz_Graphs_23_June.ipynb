```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.array([10,23,29,34,39,45,50])
y=np.array([14,21,31,43,48,55,58])
z=np.array([56,50,44,37,26,20,16])
```

```python
plt.plot(x,y);
```

```python
plt.plot(y,z);
```

```python
plt.xlabel("x values")
plt.ylabel("y values")
plt.title("line chart: x vs y")
plt.plot(x,y, marker='o',mfc='red',ls='--');
plt.grid();
```

```python
plt.xlabel("x values")
plt.ylabel("y values")
plt.title("line chart:x vs y")
plt.plot(x,y,marker='o',mfc='purple',ls='--',label='2024');
plt.plot(x,z,marker='o',mfc='yellow',ls='--',label='2025');
plt.grid();
plt.legend();####Helps distinguish between multiple plotted lines.
##Makes the graph easier to understand.
```

```python

plt.subplot(2,2,1)
plt.xlabel("x values")
plt.ylabel("y values")
plt.title("line chart: x vs y")
plt.plot(x,y, marker='o',mfc='red',ls='--');
plt.grid();

plt.subplot(2,2,2)
plt.xlabel("x values")
plt.ylabel("y values")
plt.title("line chart: x vs y")
plt.plot(x,z, marker='o',mfc='black',ls='--');
plt.grid();
```

```python
plt.figure(figsize=(16,9)) ### if we think it is zoomed in so by this we zoom out
plt.subplot(2,2,1)
plt.xlabel("x values")
plt.ylabel("y values")
plt.title("line chart: x vs y")
plt.plot(x,y, marker='o',mfc='red',ls='--');
plt.grid();

plt.subplot(2,2,2)
plt.xlabel("x values")
plt.ylabel("y values")
plt.title("line chart: x vs y")
plt.plot(x,z, marker='o',mfc='red',ls='--');
plt.grid();
```

## BAR CHART

```python
courses=['python','MIC','c++','OP','java']
students=[68,30,47,39,60]
```

```python
plt.bar(courses,students);
```

```python
# adding the labels
plt.xlabel("Subjects")
plt.ylabel("Students enrolled")
plt.title("Students enrolled in particuar subjects")
plt.bar(courses,students,color='Cyan',edgecolor='pink');
plt.grid();
```

```python
# load -->Social_Network_Ads.csv
#plot bar graph showing the count of ow many people purchesed and how many did not
# Purchased column=>0 ->Purchased, 1 -> not purchased
```

```python
Ap = pd.read_csv("Social_Network_Ads.csv")
```

```python
Ap
```

```python
counts = Ap["Purchased"].value_counts();

# Create labels
labels = ["Not Purchased", "Purchased"];

# Plot bar char
plt.bar(labels, [counts[0], counts[1]]);

plt.title("Purchase Status Count");
plt.xlabel("Status");
plt.ylabel("Number of People");
```

Scatter Chart
====

A scatter chart (scatter plot) is a type of graph used to show the relationship between two numerical variables.

It is made of many dots, where each dot represents one data point.

```python
plt.scatter(x,y,color='red');
```
