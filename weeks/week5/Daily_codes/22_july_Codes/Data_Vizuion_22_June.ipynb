```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.array([10,23,29,34,39,45,50])
y=np.array([14,21,31,43,48,55,58])
z=np.array([56,50,44,37,26,20,16])
```

Line Chart
===

```python
plt.plot(x,y);
```

```python
plt.plot(y,z);
```

```python
plt.xlabel("x values")
plt.ylabel("y values")
plt.title("line chart: x vs y")
plt.plot(x,y, marker='o',mfc='red',ls='--');
plt.grid();
# o,*,H,<,>,^
```

```python
plt.xlabel("x values")
plt.ylabel("z labels")
plt.title("line chart: x and z graph")
plt.plot(x,y, marker='*',mfc='red',ls='--');
plt.grid()
```
