```python
import numpy as np
def remove_outliers(df):
    q1=np.percentile(df['Age'],25)
    q3=np.percentile(df['Age'],75)
    iqr=q3-q1

    lb=q1-1.5*iqr
    ub=q3+1.5*iqr
    df=df[(df['Age']>=lb)& (df['Age']<=ub)]
    return df
df=remove_outliers(df)
plt.boxplot(df['Age'])
```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# 2. Load Dataset
df = sns.load_dataset("tips")

# 3. Basic Information
print(df.head())
print(df.info())
print(df.describe())
print(df.isnull().sum())
print(df.duplicated().sum())


# 4. Histogram - Total Bill
plt.figure(figsize=(6,4))
sns.histplot(df['total_bill'], bins=20, kde=True)
plt.title("Distribution of Total Bill")
plt.xlabel("Total Bill")
plt.ylabel("Frequency")
plt.show()


# 5. Histogram - Tip
plt.figure(figsize=(6,4))
sns.histplot(df['tip'], bins=20, kde=True, color='orange')
plt.title("Distribution of Tip")
plt.xlabel("Tip")
plt.ylabel("Frequency")
plt.show()


# 6. Scatter Plot - Total Bill vs Tip

plt.figure(figsize=(6,4))
sns.scatterplot(x='total_bill', y='tip', data=df)
plt.title("Total Bill vs Tip")
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.show()


# 7. Box Plot - Gender vs Tip

plt.figure(figsize=(6,4))
sns.boxplot(x='sex', y='tip', data=df)
plt.title("Tip by Gender")
plt.show()


# 8. Box Plot - Day vs Tip

plt.figure(figsize=(6,4))
sns.boxplot(x='day', y='tip', data=df)
plt.title("Tip by Day")
plt.show()


# 9. Box Plot - Smoker vs Tip

plt.figure(figsize=(6,4))
sns.boxplot(x='smoker', y='tip', data=df)
plt.title("Tip by Smoker Status")
plt.show()

# 10. Box Plot - Time vs Tip

plt.figure(figsize=(6,4))
sns.boxplot(x='time', y='tip', data=df)
plt.title("Tip by Time (Lunch vs Dinner)")
plt.show()


# 11. Box Plot - Size vs Tip

plt.figure(figsize=(6,4))
sns.boxplot(x='size', y='tip', data=df)
plt.title("Tip by Party Size")
plt.show()

# 12. Correlation Heatmap

plt.figure(figsize=(6,4))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()


# 13. Pair Plot

sns.pairplot(df, hue="smoker")
plt.show()


# 14. Feature Engineering - Tip Percentage

df["tip_pct"] = (df["tip"] / df["total_bill"]) * 100

plt.figure(figsize=(6,4))
sns.histplot(df["tip_pct"], bins=20, kde=True)
plt.title("Tip Percentage Distribution")
plt.show()
```
