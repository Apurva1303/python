```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('Social_Network_Ads.csv')
df
```

```python
#checking min and max values in age and est_salary column
```

```python
df.describe()
```

```python
plt.figure(figsize=(8,6))
plt.scatter(df['Age'], df['EstimatedSalary'], c=df['Purchased']);

plt.title("Scatter Plot of Age vs Estimated Salary")
plt.xlabel("Age")
plt.ylabel("Estimated Salary")
```

```python
#seprate the data
x=df[['Age','EstimatedSalary']]
x
```

```python
#normalization
```

```python
from sklearn.preprocessing import MinMaxScaler

sca=MinMaxScaler()
x_scaled=sca.fit_transform(x)
```

```python
x_scaled
#check min and max values of each column
```

```python
x_scaled[:,0].min(),x_scaled[:,0].max()
```

```python
x_scaled[:,1].min(),x_scaled[:,1].max()
```

```python
#plot the scatter chart to check the distribution
```

```python
plt.figure(figsize=(16,9))
plt.title('data distribution')
plt.scatter(x_scaled[:,1],x_scaled[:,0],c=df['Purchased']);
```
